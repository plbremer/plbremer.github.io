## Introduction

The [West Coast Metabolomics Center](https://metabolomics.ucdavis.edu/core-services) has collected over 64 million compound intensity measurements from over 200 thousand samples from 3 thousand studies. These stuides cover a very broad space of species, organs, and diseases. This work is a meta-analysis that:  

1) Answers broad questions like "In general, how are humans without cancer different than those with cancer?"  
2) Provides easy access to results for researchers so they can explore the compounds/metadata that interest them

## Background Knowledge - Fold Change

The only concept required to understand this analysis is a fold change. A fold change compares the average intensity of one compound in a group of samples to the average intensity of another group of samples.

For example, suppose that have two groups, where one group is labeled (Human, Lung, Healthy) and the other group is labeled (Human, Lung, Carcinoma). Then, for the example compound [alanine](https://en.wikipedia.org/wiki/Alanine), we can measure the average intensity in each group. ## Grouping Fold Changes to Answer Broad Questions

The West Coast Metabolomic Center has collected many intensity averages (over 17,000 compounds for over 1,000 metadata groups). Therefore, we can answer broad questions by compiling lists of relevant metadatagroups. We compile lists using hierarchies from literaure. We use the [ClassyFire](https://jcheminf.biomedcentral.com/articles/10.1186/s13321-016-0174-y) heirarchy for compounds, the [NCBI](https://www.ncbi.nlm.nih.gov/taxonomy) taxonomy for species, and the [MeSH](https://www.nlm.nih.gov/mesh/intro_trees.html) hierarchies for organ and disease.
