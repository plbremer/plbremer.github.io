## Background on Protein-Ligand Interactions

### Small Molecules can Interact with Certain Parts of Certain Proteins

One approach for drug development is to find small (molar mass ~<1000 grams/mol) molecules that ineract with special areas (binding sites) on proteins. These binding sites normally conduct the chemistry that is the function of the protein. By finding molecules that interact with these binding sites, the proteins are too occupied to perform chemistry on normal biomolecules.
